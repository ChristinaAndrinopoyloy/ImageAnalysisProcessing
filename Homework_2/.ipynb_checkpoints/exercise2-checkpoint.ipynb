### Part 1
The goal is to estimate the angle with respect to the horizontal axis of the billiard cue. We can use the Hough algorithm because the billiard cue is the only object with straight line. The algorithm detects straight lines, so we use it in order to detect the billiard cue. For the Hough algorith, every line is a pair of value, the rho value and the theta value. The former is the perpendicular distance from origin to the line, and the latter is the angle formed by this perpendicular line and x axis. The formula is: $\rho = x \cos \theta + y \sin \theta$. 

The input of the Hough algorithm is a binary image, so we apply the Canny algorithm and then give the image as input to the Hough algorithm. The last argument is relative to the size of the line, so we give a large value because the billiard cue is quite long. Then, we plot the lines that the algorith found. 

The function below contains the aforementioned procedures. Also, it includes another one step, the first one the is the rotation of the image. The rotation of the image will help us to determine the angle between the billiard cue and the x axis. So, the function *hough_pipeline*:
- rotates the input image
- converts it to grayscale
- executes the Canny algorithm
- executes the Hough algorithm for finding lines
- returns the image and the lines

```python
def hough_pipeline(img, rotation=20):
    image = ndimage.rotate(img, rotation, reshape=False)
    gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray,50,150,apertureSize = 3)
    lines = cv2.HoughLines(edges,1,np.pi/180,200)
    for line in lines:
        for rho,theta in line:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))
            cv2.line(image,(x1,y1),(x2,y2),(255,0,0),2)
    return image, lines
```

For finding the angle between the x axis and the billiard cue, we use the formula below that is implemented by the function *degree_calc*:
\begin{align}
a = (\theta + 90) mod 180
\end{align}

where $\theta$ is the angle between the x axis and the vector that is perpendicular to the billiard cue. For the calculation of the angle between the x axis and the billiard cue, we can use the triangle that is created by the vector, the billiard cue and the x axis.

<img src="image31_triangle.png">

The triangle is right, so the angle under the x axis is equal to $90 - \theta$ and the complementary angle is equal to $90 + \theta$. 

If the direction of the vector shows the South, the angle between the line and the x axis is equal to 180 and if the vector shows the North, the angle is equal to 0. However, the two lines are parallel. Because 0 is equivalent of 180, we use the mod in the first formula. 

```python
def degree_calc(lines):
    theta = lines[0][0][1]
    return (round(math.degrees(theta))+90)%180
```

We rotate the initial image using some random values in order to estimate the rotation angle for image so the billiard cue will be parallel to the x axis.

```python
img = cv2.imread('image31.png')

fig, axes = plt.subplots(1, 4, figsize=(26,8))
plt.subplot(141), plt.imshow(img), plt.axis('off') 
plt.title('Initial Image')

image = ndimage.rotate(img, 40, reshape=False)
plt.subplot(142), plt.imshow(image), plt.axis('off') 
plt.title('Rotate the image: 40')

image = ndimage.rotate(img, -40, reshape=False)
plt.subplot(143), plt.imshow(image), plt.axis('off') 
plt.title('Rotate the image: -40')

image = ndimage.rotate(img, -70, reshape=False)
plt.subplot(144), plt.imshow(image), plt.axis('off') 
plt.title('Rotate the image: -70')

plt.show()
```

From the figures below we can understand that the angle that we search is between the values (-70,-40) and we execute the *hough_pipeline* for different rotation values in this range. The *hough_pipeline* returns a line that corresponds to the billiard cue and then we calculate the angle between the billiard cue and the x axis. If this angle is 0 than we found the optimal value for the rotation of the image (so now the billiar cue is parallel to the x axis) and this rotation corresponds to the angle of the billiard with respect to the horizontal axis.

```python
optimal_rotation = 0
for rotation in range(-60,-40,3):
    image, lines = hough_pipeline(img, rotation)
    degree_value = degree_calc(lines)
    if degree_value == 0:
        optimal_rotation = rotation   


image, lines = hough_pipeline(img, rotation=optimal_rotation)
plt.imshow(image), plt.axis('off') 
plt.title("The angle between the billiard cue and the horizontal axis is "+str(optimal_rotation))
plt.show()
```

```python
image = copy.copy(blur)

# step 1:        
detected_circles = cv2.HoughCircles(image, cv2.HOUGH_GRADIENT, 1, 20, param1 = 25, param2 = 25, minRadius = 20, maxRadius = 40)
# Draw circles that are detected.
if detected_circles is not None:
    detected_circles = np.uint16(np.around(detected_circles))
    for pt in detected_circles[0, :]:
        a, b, r = pt[0], pt[1], pt[2]
        cv2.circle(image, (a, b), r, (255, 255, 255), 2)
        
# step 2:
black = np.zeros(image.shape)
for x, y, r in detected_circles[0]:
    cv2.circle(black, (x,y), int(r+2), 255, -1)  # -1 to draw filled circles 
    
# step 3 A:
bytemask = np.asarray(black*255, dtype=np.uint8)
inpainted = cv2.inpaint(image, bytemask, inpaintRadius=10, flags=cv2.INPAINT_TELEA)
# step 3 B:
image[black == 255] = image.mean()

# plot
fig, axes = plt.subplots(1, 4, figsize=(18,8))
plt.subplot(141), plt.imshow(image, cmap='gray'), plt.axis('off'), plt.title('HoughCircles')
plt.subplot(142), plt.imshow(black.astype('uint8'), cmap='gray'), plt.axis('off'), plt.title('Mask')
plt.subplot(143), plt.imshow(inpainted.astype('uint8'), cmap='gray'), plt.axis('off'), plt.title('Remove cycles (A)')
plt.subplot(144), plt.imshow(image.astype('uint8'), cmap='gray'), plt.axis('off'), plt.title('Remove cycles (B)')
plt.show()
detected_circles = []
```

```python
# A
rotated_image = ndimage.rotate(inpainted, -51, reshape=True)
blur = cv2.GaussianBlur(rotated_image,(7,7),0)
edges = cv2.Canny(blur,50,150,apertureSize = 3)
edges = np.vstack((edges, np.zeros((1,edges.shape[1])))) # add a row, so the two images have the same size

fig, axes = plt.subplots(1, 2, figsize=(12,5))
plt.subplot(121), plt.imshow(rotated_image, cmap='gray'), plt.axis('off'), plt.title('Rotated Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.axis('off'), plt.title('Canny')
plt.show()

# B
rotated_image = ndimage.rotate(image, -51, reshape=True)
blur = cv2.GaussianBlur(rotated_image,(7,7),0)
edges = cv2.Canny(blur,50,150,apertureSize = 3)
edges = np.vstack((edges, np.zeros((1,edges.shape[1])))) # add a row, so the two images have the same size

fig, axes = plt.subplots(1, 2, figsize=(12,5))
plt.subplot(121), plt.imshow(rotated_image, cmap='gray'), plt.axis('off'), plt.title('Rotated Image')
plt.subplot(122), plt.imshow(edges, cmap='gray'), plt.axis('off'), plt.title('Canny')
plt.show()
```

```python
kernel = np.ones((20,5),np.uint8)
closing = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)
plt.imshow(closing, cmap='gray'), plt.axis('off'), plt.title('Closing')
plt.show()
```

```python
kernel = np.ones((3,24),np.uint8)
opening = cv2.morphologyEx(closing, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap='gray'), plt.axis('off'), plt.title('Opening')
plt.show()
```

```python
init_im_2 = initial_image2.copy()

fig, axes = plt.subplots(1, 4, figsize=(30,12))
plt.subplot(141), plt.imshow(opening, cmap='gray'), plt.axis('off'), plt.title('Mask')
plt.subplot(142), plt.imshow(initial_image2, cmap='gray'), plt.axis('off'), plt.title('Initial Image 2')
init_im_2[opening == 0] = (0,0,0)

plt.subplot(143), plt.imshow(init_im_2), plt.axis('off'), plt.title('Final Result of the Pipeline') 
plt.subplot(144), plt.imshow(target_image, cmap='gray'), plt.axis('off'), plt.title('Target Image')
plt.show()
```
